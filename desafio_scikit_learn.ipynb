```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dados_vendas = {
    'mes': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
    'vendas': [2000, 2200, 2300, 2500, 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300]
}
```

```python
df_vendas = pd.DataFrame(dados_vendas)
```

```python
df_vendas
```

```python
df_vendas['mes_num'] = df_vendas.index + 1
```

```python
df_vendas
```

```python
X = df_vendas[['mes_num']]
y = df_vendas['vendas']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=51)
```

```python
model = LinearRegression().fit(X_train, y_train)
```

```python
dezembro = np.array([[12]])
venda_dezembro = model.predict(dezembro)
print(f'Previsão de vendas para Dezembro: {venda_dezembro[0]}')
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df_vendas['vendas'], bins=10, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histograma das Vendas')
plt.xlabel('Vendas')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df_vendas['mes_num'], df_vendas['vendas'], color='blue')
plt.plot(df_vendas['mes_num'], model.predict(df_vendas[['mes_num']]), color='red')
plt.title('Vendas por Mês')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.xticks(ticks=df_vendas['mes_num'], labels=df_vendas['mes'], rotation=45)
plt.show()
```
